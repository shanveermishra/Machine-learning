```python
import pandas as pd 
from sklearn.model_selection import train_test_split
```

Load The Dataset & Understanding the Dataset

```python
df = pd.read_csv(r"C:\Users\Asus\Downloads\500hits.csv"  , encoding = 'latin-1')
df.head() #
```

Creating X and Y Variables

```python
X = df.drop(columns=['PLAYER' ,'HOF']) # Player column utna jruri nahi features me target predict krne ke liye to hata denge aur HOF target hai toh uss se X se hata denge 
X.head()
```

```python
y = df['HOF'] 
y.head()
```

Implementing Train test Split

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , random_state=11 , test_size= 0.2)
# random_state ek fixed seed hai jo ensure karta hai ki same code dobara run karne par data har baar same tarike se train aur test me split ho,
#Agar kissi ne same dataset + same code + same random_state=11 use kiya â exact same train-test split milega 
#test_size define karta hai ki data ka kitna portion model ko test karne ke liye alag rakha jayega ,
# 0.2 = 20% data â test set , 80% data â train set
```
