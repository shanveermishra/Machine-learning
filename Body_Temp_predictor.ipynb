# Body Temperature Prediction While Working Out in Gym using Linear Regression

## Description 
### A beginner machine learning project that predicts body temperature based on workout duration using Simple Linear Regression.

```python
import pandas as pd 
import matplotlib.pyplot as plot 
```

```python
df = pd.read_csv(r'D:\Jupyter Lab\archive\exercise.csv') 
```

```python
df.head()
```

```python
new_df = df.drop(['User_ID' , 'Gender' , 'Age' , 'Height' ,'Weight' , 'Heart_Rate'],axis=1) #inplace=True -> OG dataset ko bhi aise hi save kar dega 
```

```python
new_df.info()
new_df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

Create X and y

```python
X = new_df[['Duration']]  
y = new_df['Body_Temp']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X , y , random_state=10 ,test_size=0.2 )
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr = LinearRegression()
```

```python

lr.fit(X_train,y_train)  
```

```python
X_train.shape 
```

```python
y_train.shape
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))  
```

```python
plot.scatter(new_df['Duration'] , new_df['Body_Temp'])
plot.xlabel('Duration')
plot.ylabel('Body_Temp')
plot.plot(X_train,lr.predict(X_train) , color = 'red')
```

# Output 
### The model successfully learned the relationship between workout duration and body temperature.
